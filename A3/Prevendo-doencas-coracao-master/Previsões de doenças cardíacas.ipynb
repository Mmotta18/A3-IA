### Processamento de Dados

Depois de explorar o conjunto de dados, observei que preciso converter algumas variáveis categóricas em variáveis fictícias e dimensionar todos os valores antes de treinar os modelos de Machine Learning.
Primeiro, usarei o método 'get_dummies' para criar colunas fictícias para variáveis categóricas.

```python
dataset = pd.get_dummies(df, columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
dataset[columns_to_scale] = standardScaler.fit_transform(dataset[columns_to_scale])
```

```python
dataset.head()
```

```python
y = dataset['target']
X = dataset.drop(['target'], axis = 1)
```

```python
from sklearn.model_selection import cross_val_score
knn_scores = []
for k in range(1,21):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score=cross_val_score(knn_classifier,X,y,cv=10)
    knn_scores.append(score.mean())
```

```python
plt.plot([k for k in range(1, 21)], knn_scores, color = 'red')
for i in range(1,21):
    plt.text(i, knn_scores[i-1], (i, knn_scores[i-1]))
plt.xticks([i for i in range(1, 21)])
plt.xlabel('Número de Vizinhos (K)')
plt.ylabel('Pontos')
plt.title('Pontuações do Classificador de Vizinhos K para diferentes valores de K')
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 12)
score=cross_val_score(knn_classifier,X,y,cv=10)
```

```python
score.mean()
```
